```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

| Feature                 | Typical Range  | Low Value Means                  | Medium Value Means  | High Value Means                |
| ----------------------- | -------------- | -------------------------------- | ------------------- | ------------------------------- |
| **Variance**            | `0 → ~10,000+` | Very smooth / flat image         | Moderate contrast   | High contrast, many edges       |
| **Skewness**            | `-3 → +3`      | Bright-heavy image (overexposed) | Balanced brightness | Dark-heavy image (underexposed) |
| **Kurtosis** *(excess)* | `-2 → +10+`    | Flat distribution, few edges     | Natural image       | Sharp peaks, strong edges       |
| **Entropy**             | `0 → 8`        | Very simple / uniform            | Moderate complexity | Highly complex / detailed       |
